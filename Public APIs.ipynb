```python
import pandas as pd
import requests
```

# Last FM

https://www.last.fm/api/show/tag.getTopTracks

```python
import getpass

apikey = getpass.getpass()
```

```python
base_url = "http://ws.audioscrobbler.com/2.0/"
```

```python
toptracks_url = base_url + "?method=tag.gettoptracks&tag=disco&api_key=" + apikey + "&format=json"
    
toptracks_response = requests.get(toptracks_url)
# toptracks_response.json()
```

```python
# Extracting array on tracks from the original json

results_top10 = toptracks_response.json()

tracks = results_top10["tracks"]
track = tracks["track"]
```

```python
from pandas.io.json import json_normalize

toptrack_results = pd.json_normalize(track)
toptrack_results.head()
```

```python
# Deleting some unwanted columns

toptrack_results = toptrack_results.drop(["mbid", "image", "streamable.#text", "streamable.fulltrack", "artist.mbid"], axis = 1)
toptrack_results.head()
```

# UK Bank Holidays

```python
holiday_url = "https://www.gov.uk/bank-holidays.json"

holiday_response = requests.get(holiday_url)

holiday_json = holiday_response.json()
# holiday_json
```

```python
records = holiday_json["england-and-wales"]
england_and_wales = records["events"]

england_and_wales_df = pd.json_normalize(england_and_wales)
england_and_wales_df["division"] = "england-and-wales"
england_and_wales_df
```

```python
record1 = holiday_json["scotland"]
scotland = record1["events"]

scotland_df = pd.json_normalize(scotland)
scotland_df["division"] = "scotland"
scotland_df
```

```python
record1 = holiday_json["northern-ireland"]
northern_ireland = record1["events"]

northern_ireland_df = pd.json_normalize(northern_ireland)
northern_ireland_df["division"] = "northern-ireland"
northern_ireland_df
```

```python
holiday_df = pd.concat([england_and_wales_df, scotland_df ,northern_ireland_df], axis = 0)
holiday_df
```
